```python
import pandas as pd

class electronicModelBased_RecommenderAgent:
    
    def load_products(self, csv_file_path):
        print("Load Electronic Product \n")        
                
        df = pd.read_csv(csv_file_path, encoding= "unicode_escape")
        
        self.df = df
        
        print(df.head())
        
 #adds preference to what electronic user would be interested in purchasing based on their budget   
    def add_preference(self):
        print('\n To Recommend Electronics Products to Purchase Based on User Preferences \n')
        
        pref_store = str(input("Store: Walmart, Best Buy, ..., ..., Target, Gamestop: "))
        pref_budget = float(input("Budget: $"))
        
        self.pref_store = pref_store
        self.pref_budget = pref_budget     
            
 #displays electronic item based on preference        
    def recommend_product(self):
        print("\n Electronic Item Recommendation Based on your Preferences")        
                
        pref_store = self.pref_store
        pref_budget = self.pref_budget
        df = self.df       
        
        filtered_df = df[(df['Store'] == pref_store) & (df['Price'] < pref_budget)]
        
        cumulative_budget = 0.0
        displayed_rows = []

        # Iterate through the DataFrame and accumulate rows until the budget is reached
        for index, row in filtered_df.iterrows():
            cumulative_budget = cumulative_budget + row['Price']
            if cumulative_budget <= pref_budget:
                displayed_rows.append(row)
            else:
                break        
        
        recommended_prods = pd.DataFrame(displayed_rows)
        
        print(recommended_prods)
                    
            
                 
                
                
                
                
                        
#----------------------------Class Driver --------------------------------------    
agent = electronicModelBased_RecommenderAgent()

csv_file_path = 'ElectronicProduct.csv'

# Agent designed as semi-autonomous 
while True:    

    agent.load_products(csv_file_path)

    agent.add_preference()

    agent.recommend_product()

    n = input("\n Please press Enter-key to continue or enter 'q' to Quit: ")
    if n.strip() == 'q':
        break
```
